```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
X = [n for n in range(1,100)]
X_r = np.arange(1, 100, 0.001)
Y = np.sin(X)
Y_r = np.sin(X_r)
```

```python
Y[:5]
```

```python
plt.figure(figsize=(24,9))
plt.scatter(X,Y,s=20,c="red",label="sin(n)")
plt.plot(X_r,Y_r,color='green',label="sin(x)")
plt.vlines(X, 0, Y, color='purple', linewidth=1)
plt.title("Plot of sequence sin(n) and function sin(x) upto n = 100")
plt.xlabel("n")
plt.ylabel("sin(n)")
plt.axhline(y=0, color="black")
plt.axhline(y=1, color="black")
plt.axhline(y=-1, color="black")
plt.legend()
plt.show()
```

```python
X = [n for n in range(1,1000)]
Y = np.sin(X)
plt.figure(figsize=(24,9))
plt.scatter(X,Y,s=20,c="red",label="sin(n)")
plt.title("Plot of sequence sin(n) upto n = 1000")
plt.show()
```

```python
X = [n for n in range(1,10000)]
Y = np.cos(X)
plt.figure(figsize=(24,9))
plt.scatter(X,Y,s=20,c="red",label="cos(n)")
plt.title("Plot of sequence cos(n) upto n = 10000")
plt.show()
```

```python
X = [n for n in range(1,101)]
Y = []
X_r = np.arange(1,101,0.001)
Y_r = []
for x in X_r:
    y = (1 + 1/x) **x
    Y_r.append(y)
for x in X:
    y = (1 + 1/x) **x
    Y.append(y)
plt.figure(figsize=(16,9))
plt.scatter(X,Y,c='red',s=50,label="sequence")
plt.plot(X_r,Y_r,color="green",label="real function")
plt.title(r"$x_n = \left(1 + \frac{1}{n}\right)^n$")
plt.axhline(y=np.exp(1),color="black",label="y=e")
plt.legend()
plt.show()
```
